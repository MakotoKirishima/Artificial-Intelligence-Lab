```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
data = {
  'hari_aktif': [5, 10, 15, 20, 25, 30, 35, 40, 45, 50],
  'clear_drops': [900, 1850, 2700, 3800, 4600, 5500, 6400, 7200, 8100, 9050]
}

df = pd.DataFrame(data)
print(df)
```

```python
x = df[['hari_aktif']]
y = df ['clear_drops']
```

```python
x_train, x_test, y_train, y_test = train_test_split(
    x,y,test_size=0.2, random_state=42
)
```

```python
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
print("Intercept (a):", model.intercept_)
print("Coefficient (b):", model.coef_[0])
```

```python
y_pred = model.predict(x_test)

print("Data Testing:")
print(x_test)
print("clear_drops asli:", y_test.values)
print("clear_drops prediksi:", y_pred)
```

```python
print("MAE:", metrics.mean_absolute_error(y_test, y_pred))
print("MSE:", metrics.mean_squared_error(y_test, y_pred))
print("RMSE:", np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```

```python
plt.scatter(x, y, label="Data Asli")
plt.plot(x, model.predict(x), label="Garis Regresi")
plt.xlabel("Hari Aktif Bermain")
plt.ylabel("Jumlah Clear Drops")
plt.title("Regresi Linear: Hari Aktif vs Clear Drops")
plt.legend()
plt.show()
```

Analisis:
* Model menunjukkan korelasi positif yang sangat kuat antara hari aktif bermain dan jumlah Clear Drops.
* Nilai Coefficient (b) sebesar 180.25  yang berarti kalau setiap penambahan 1 hari aktif bermain, jumlah Clear Drops diperkirakan bertambah sekitar 180 unit.
* Nilai Error (MAE: 43.75, RMSE: 52.78) tergolong sangat kecil dibandingkan skala data target (ribuan), membuktikan bahwa garis regresi cocok (fit) dengan pola data asli.
